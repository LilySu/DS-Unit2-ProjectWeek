```python
X = df.drop(columns = ['Shipping Discount','Order Type','Discount Amount','Number of Items','Sale Date', 'Date Shipped','Coupon Code', 'Coupon Details','Adjusted Net Order Amount','InPerson Discount','InPerson Location', 'Status', 'Adjusted Order Total', 'Adjusted Card Processing Fees','Order Total','Order Value','Shipping', 'Sales Tax', 'Order Net', 'Card Processing Fees'])
y = df['Order Total']


import category_encoders as ce
encoder = ce.OrdinalEncoder()
X = encoder.fit_transform(X)
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
  X, y, train_size = 0.8, test_size=0.2, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(
  X_train, y_train, train_size = 0.8, test_size=0.2, random_state=42)

X_train.shape, X_val.shape, X_test.shape, y_train.shape, y_val.shape, y_test.shape
```

```python
import category_encoders as ce
from sklearn.ensemble import RandomForestRegressor

encoder = ce.OrdinalEncoder()
# X_train = encoder.fit_transform(X_train)
# X_val = encoder.transform(X_val)
# X_test = encoder.transform(X_test)

model = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_val)
y_pred
```

```python
results = pd.DataFrame({'y_val': y_val, 'y_pred': y_pred})
top2000 = results.sort_values(by='y_pred', ascending=False)[:500]
top2000.sample(n=50)
```

```python
top2000[:10].plot.barh(figsize=(10,20))
```

```python
from sklearn.pipeline import make_pipeline
pipeline = make_pipeline(
    ce.OrdinalEncoder(), 
    RandomForestRegressor(n_estimators=1000, random_state=42, n_jobs=-1)
)

# Fit
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_val)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.figure(figsize=(10,10))
rf = pipeline.named_steps['randomforestregressor']
importances = pd.Series(rf.feature_importances_, X_train.columns)
x = np.arange(10)
ys = [i+x+(i*x)**2 for i in range(10)]
colors = cm.rainbow(np.linspace(0, 1, len(ys)))
importances.sort_values().plot.barh(color=colors);
```

```python
from pdpbox.pdp import pdp_interact, pdp_interact_plot

features = ['Ship State', 'Order Total']

interaction = pdp_interact(
    model=model, 
    dataset=X_val, 
    model_features=X_val.columns, 
    features=X_train.columns
)

pdp_interact_plot(interaction, plot_type='grid', feature_names=features);
```
